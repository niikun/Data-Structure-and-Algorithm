# Big-O Notation:Plot
このノートブックの目的は、アルゴリズム分析で頻繁に使われるいくつかの関数の成長順序を視覚化することである。このノートブックはインタラクティブなノートブックであることに注意してください。好きな関数を入れたり、下の範囲を変更してみて、何が起こるか見てください。

## Definitions
まず定義を思い出すことから始めよう。$f(n)$ と $g(n)$ という2つの関数を考える。

- $f$ が $g$ よりゆっくり大きくなる時、$f \prec g$ と書く。  
- $f$ が $g$ より早く大きくならない時、$f \preceq g$ と書く。  
- 定数 $c$ があり、$f(n) \leq c \cdot g(n)$
3つの重要な注意
1. $f \prec g$は$f=o(g)$(small-o)と同じで、$f \preceq g$は$f=O(g)$ (big-O)と同じ。このノートでは、$⪯$表記にこだわることにした。
多くの学習者は" $5n^2=O(n^3)$について、「これは間違っている！5n2=O(n2)が正しい !」という。実は、これらの記述はどちらも正しいく、$5n^2=O(n^3)$であり、また$5n^2=O(n^2)$でもある。
この２つの式は$5n^2$は$n^2$と$n^3$より早く大きくならないと言っているだけである。
$⪯$記法では、次のように表現される：$5n^2 \preceq n^2$ and $5n^2 \preceq n^3$.   
同じように:  $x=2$, のとき $x \le 2$ または $x \le 3$  
2.  $f \prec g$, のとき $f \preceq g$。つまり、もし、 $f$ が $g$よりゆっくり大きくなる時 $f$は $g$は早く大きくならないということである。  
3.  標準的な不等号 $\le$ の代わりに $\preceq$ という記号を使う必要がある： すべての$n$に対して$f(n) \le g(n)$のとき、$f$は$\le g$である。また、$5n^2 \not \le n^2$だが、 $5n^2 \preceq n^2$は成立する

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

$7n^2+6n+5$
$1≤n≤100$

```python
n = np.linspace(1,100)
y = 7*n**2 + 6*n + 5
plt.plot(n,y)
plt.show()
```

$20n$と比較してみよう.
$20n$は、$7n^2+6n+5$ より早く大きくならない

```python
n=np.linspace(1,100)
plt.plot(n,n*20,label="$20n$")
plt.plot(n,7*n**2+6*n+5,label="$7n^2+6n+5$")
plt.legend()
plt.show()
```

## Common rules
可視化を進める前に、アルゴリズム分析でよく出てくる関数の成長順序を比較する一般的なルールを復習しておこう。  
1. 乗法定数は省略できる：$c⋅f⪯f$
. Examples: $5n^2⪯n^2$
,  $\frac{n^2}{3}⪯n^2$  
2. 2つの多項式のうち、次数の大きい方が早く成長する：$n^a⪯n^b$ for $0≤a≤b$
例えば: $n≺n^2$
, $n2≺n3$
, $n^0≺\sqrt{n}$
3. 任意の対数関数は任意の多項式関数よりも成長が遅い。 $ (logn)a≺n^b$
  for  $a,b>0$  例：$n^3≺2^n$
 , $n^{10}≺1.1^n$

4. 任意の多対数関数は任意の多項式関数よりも成長が遅い。$(logn)a≺n^b$  for $a,b>0$ 例:  $(logn)^3≺\sqrt{n}, nlogn≺n^2$
.
4. 小さな次数は省略できる。 もし $f+g⪯g$　なら、$f≺g$。

小さな次数は省略できる

```python
n = np.linspace(1,100)
plt.plot(n,7*n**2+6*n+5,label="$7n^2+6n+5$")
plt.plot(n,7*n**2,label="$7n^2$")
plt.legend()
plt.show()
```

### Rule 1: Multiplicative constants can be ommitted  
$\frac{7n^2+6n+5}{n^2}$

```python
n = np.linspace(1, 100)
plt.plot(n, (7 * n * n + 6 * n + 5)/(n * n))
plt.show()
```

### Rule 2: Out of two polynomials, the one with larger degree grows faster  
For constants  $a>b>0$
 ,  $n^a$
  grows faster than  $n^b$
 . This, in particular, means that  $n^b=O(na)$
 .   
 To visualize it, let's plot  n
 ,  $n^2$
 , and  $n^3$
 .

```python
n = np.linspace(1,10)
plt.plot(n,n,label="n")
plt.plot(n,n**2,label="$n^2$")
plt.plot(n,n**3,label="$n^3$")
plt.legend()
plt.show()
```

### Rule 3: Any polynomial grows slower than any exponentia  
  $n^4$
  and  $2^n$
  in the range  $1≤n≤10$
 .

```python
n = np.linspace(1,10)
plt.plot(n,n**4,label="$n^4$")
plt.plot(n,2**n,label="$2^n$")
plt.title("$1≤n≤10$")
plt.legend()
plt.show()
```

```python
n = np.linspace(1,20)
plt.plot(n,n**4,label="$n^4$")
plt.plot(n,2**n,label="$2^n$")
plt.title("$1≤n≤20$")
plt.legend()
plt.show()
```

### Rule 4: Any polylogarithm grows slower than any polynomial  
$logn$  and $n$

```python
n = np.linspace(1,20)
plt.plot(n,np.log(n),label="$logn$")
plt.plot(n,n,label="$n$")
plt.title("$1≤n≤20$")
plt.legend()
plt.show()
```

 $ (logn)^3$  versus  $\sqrt{n}$
  

```python
n = np.linspace(1,100)
plt.plot(n,(np.log(n))**3,label="$(logn)^3$")
plt.plot(n,n**0.5,label="$\sqrt{n}$")
plt.title("$1≤n≤1000$")
plt.legend()
plt.show()
```

```python
n = np.linspace(1,10**6)
plt.plot(n,(np.log(n))**3,label="$(logn)^3$")
plt.plot(n,n**0.5,label="$\sqrt{n}$")
plt.title("$1≤n≤10**6$")
plt.legend()
plt.show()
```

驚くべきことに、対数関数は依然として多項式関数を上回っている！このことは、それほど大きくない$n$の値に対して、どちらの関数がどのように振る舞うかを見ただけで、どちらの関数が速く成長するかを決めるのは危険であることを示している。
 . どんな多項式もどんな多項式より速く成長する」という規則は、最終的に多項式関数が多項式関数よりどんどん大きくなることを意味する。しかし、この規則では、$n$  のどのような値で初めてそうなるかは明示していない。

```python
n = np.linspace(1,10**8)
plt.plot(n,(np.log(n))**3,label="$(logn)^3$")
plt.plot(n,n**0.5,label="$\sqrt{n}$")
plt.title("$1≤n≤10**8$")
plt.legend()
plt.show()
```

```python
n = np.linspace(1, 100)
plt.plot(n, n ** .1, label="$n^{0.1}$")
plt.plot(n, np.log(n) ** 5, label="$(log n)^5$")
plt.legend(loc='upper left')
plt.show()
```
